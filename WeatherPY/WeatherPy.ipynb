## Three Observations

1.Latitude doesnt seem to have any effect at all on humidity in either hemisphere.
2.Latitude doesnt seem to have any effect at all on cloudiness in either hemisphere.
3.Latitude does have an effect on temperature though showing that as cities move further away from the equator they vbecome colder.


```python


import json
import requests
from api_keys import weather_api_key
import pandas as pd
import numpy as np
import random
from pprint import pprint
from citipy import citipy
import matplotlib.pyplot as plt
import scipy.stats as st

print(weather_api_key)
```

## Generate Cities List

```python
lng=[]
lat=[]

for a in range(1500):
    lng.append(np.random.uniform(-180,180))
    lat.append(np.random.uniform(-90,90))

lat_long_dc={'lat':lat,'long':lng}
lat_long_df=pd.DataFrame.from_dict(lat_long_dc)


cities_name=[]
countrycode=[]

for x in range(0,1500):
    latitude=lat_long_df.iloc[x]['lat']
    longitude=lat_long_df.iloc[x]["long"]
    city=citipy.nearest_city(latitude,longitude)
    cities_name.append(city.city_name)
    countrycode.append(city.country_code)
    

cities_info=pd.DataFrame({ "City Name":cities_name,"Country":countrycode})

cities_info.head()

    

```

```python
cities_info = cities_info.drop_duplicates("City Name", keep='first')
cities_info["City Name"].count()
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
cities_info["Temp (F)"]=""
cities_info["Humidity %"]=""
cities_info["Cloudiness %"]=""
cities_info["Wind Speed (mph)"]=""
cities_info['lat']=""
cities_info['long']=""
cities_info
```

```python
url = "https://api.openweathermap.org/data/2.5/weather?q={},{}&appid=c4d790a751167825c1dd34dc592ab9b1&units=imperial"

cities_info = cities_info
data_num=1

for index, row in cities_info.iterrows():
    try:
        city_url=url.format(row["City Name"],row["Country"])
        city_response=requests.get(city_url).json()
        cities_info.loc[index, "Temp (F)"]=city_response['main']['temp_max']
        cities_info.loc[index, "Humidity %"]=city_response['main']['humidity']
        cities_info.loc[index, "Cloudiness %"]=city_response['clouds']['all']
        cities_info.loc[index, "Wind Speed (mph)"]=city_response['wind']['speed']
        cities_info.loc[index, "lat"]=city_response['coord']['lat']
        cities_info.loc[index, "long"]=city_response['coord']['lon'] 
        print(f'Getting data from data id {data_num} which is {row["City Name"]}')
        data_num=data_num+1
    except KeyError:
        print(f'There was a problem with {row["City Name"]} and we are skiping.')
        cities_info.loc[index, "Temp (F)"]="NaN"
        


    
```

### Convert Raw Data to DataFrame

## Linear Regression

```python
nhemi=cities_info.loc[cities_info['lat'] >0]
shemi=cities_info.loc[cities_info['lat'] <0]

nhemi["Temp (F)"] = nhemi["Temp (F)"].astype(float)
shemi["Temp (F)"] = shemi["Temp (F)"].astype(float)
nhemi["Humidity %"] = nhemi["Humidity %"].astype(float)
shemi["Humidity %"] = shemi["Humidity %"].astype(float)
nhemi["Cloudiness %"] = nhemi["Cloudiness %"].astype(float)
shemi["Cloudiness %"] = shemi["Cloudiness %"].astype(float)
nhemi["Wind Speed (mph)"] = nhemi["Wind Speed (mph)"].astype(float)
shemi["Wind Speed (mph)"] = shemi["Wind Speed (mph)"].astype(float)
```

```python
nhemi['lat']=nhemi['lat'].astype(float)
shemi['lat']=shemi['lat'].astype(float)
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python

(slope,intercept,rvalue,pvalue,stderr)=st.linregress(nhemi["lat"],nhemi["Temp (F)"])

regress_values = slope*nhemi["lat"]+intercept

line_eq = "y=" + str(round(slope,2)) + "x + " + str(round(intercept,2))


plt.scatter(nhemi["lat"],nhemi["Temp (F)"])
plt.title("North Hemisphere City Latitude vs Max Temp (F) on 09/22/2020")
plt.ylabel("Max Temp (F)")
plt.xlabel("Latitude")
plt.grid(True)

plt.plot(nhemi["lat"],regress_values,"r")
plt.annotate(line_eq,(0,0),fontsize=15,color="r")

cc=round(rvalue,2)

print(f'The corrleation coefficient is {cc} meaning there is a strong negative corelation.')
print(f'The Linear Regress model for this is {line_eq}')

plt.savefig("Temp_vs_NHempLat.png")
```
